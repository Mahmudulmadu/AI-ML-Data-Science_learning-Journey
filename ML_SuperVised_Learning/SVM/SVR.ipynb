```python
from sklearn.svm import SVR
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import datasets
```

```python
df = datasets.load_diabetes(as_frame=True).frame
```

```python
df.head()
```

```python
X = df.drop("target", axis=1)
y = df["target"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
y_train_scaled = scaler.fit_transform(y_train.values.reshape(-1, 1)).ravel()
y_test_scaled = scaler.transform(y_test.values.reshape(-1, 1)).ravel()
```

```python
from sklearn.metrics import r2_score
model = SVR()
model.fit(X_train, y_train_scaled)
y_test_pred_scaled = model.predict(X_test)
y_train_pred_scaled = model.predict(X_train)

print("r2 Score of Train : ", r2_score(y_train_scaled,y_train_pred_scaled)  )
print("r2 Score of Test : ", r2_score(y_test_scaled,y_test_pred_scaled)  )
```

```python
#Linear
from sklearn.metrics import r2_score
model = SVR(kernel="linear")
model.fit(X_train, y_train_scaled)
y_test_pred_scaled = model.predict(X_test)
y_train_pred_scaled = model.predict(X_train)

print("r2 Score of Train : ", r2_score(y_train_scaled,y_train_pred_scaled)  )
print("r2 Score of Test : ", r2_score(y_test_scaled,y_test_pred_scaled)  )
```

```python
#Poly
from sklearn.metrics import r2_score
model = SVR(kernel="poly")
model.fit(X_train, y_train_scaled)
y_test_pred_scaled = model.predict(X_test)
y_train_pred_scaled = model.predict(X_train)

print("r2 Score of Train : ", r2_score(y_train_scaled,y_train_pred_scaled)  )
print("r2 Score of Test : ", r2_score(y_test_scaled,y_test_pred_scaled)  )
```

```python
#sigmoid
from sklearn.metrics import r2_score
model = SVR(kernel="sigmoid")
model.fit(X_train, y_train_scaled)
y_test_pred_scaled = model.predict(X_test)
y_train_pred_scaled = model.predict(X_train)

print("r2 Score of Train : ", r2_score(y_train_scaled,y_train_pred_scaled)  )
print("r2 Score of Test : ", r2_score(y_test_scaled,y_test_pred_scaled)  )
```

# HyperParameter tuning using GridSearchCV


```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    "C": [1,2,5,10,50,100],
    "kernel" : ["rbf", "linear"],
    "epsilon" : [0.001, 0.01, 0.02, 0.3, 0.5]
}
```

```python
svr = SVR()

grid_search = GridSearchCV(svr, param_grid, scoring="r2", cv=5)
```

```python
grid_search.fit(X_train, y_train_scaled)
```

```python
print("best params - ", grid_search.best_params_)
```

```python
best_model = SVR(kernel="linear", C=50, epsilon=0.3)
best_model.fit(X_train, y_train_scaled)
y_test_pred_scaled = best_model.predict(X_test)
y_train_pred_scaled = best_model.predict(X_train)

print("r2 Score of Train : ", r2_score(y_train_scaled,y_train_pred_scaled)  )
print("r2 Score of Test : ", r2_score(y_test_scaled,y_test_pred_scaled)  )
```

```python
# Linear SVR

from sklearn.svm import LinearSVR
model = LinearSVR(C=50, epsilon=0.3, max_iter=5000)
model.fit(X_train, y_train_scaled)
y_test_pred_scaled = model.predict(X_test)
y_train_pred_scaled = model.predict(X_train)

print("r2 Score of Train : ", r2_score(y_train_scaled,y_train_pred_scaled)  )
print("r2 Score of Test : ", r2_score(y_test_scaled,y_test_pred_scaled)  )
```
