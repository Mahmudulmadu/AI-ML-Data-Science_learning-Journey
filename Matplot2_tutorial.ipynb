```python
import matplotlib.pyplot as plt
```

```python
# Movies Data

# Oscar Winner Movies
oscar_movies = [
    "The Dark Knight", 
    "The Hurt Locker",  
    "The King's Speech", 
    "The Artist",     
    "Argo"
]
oscar_years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M


plt.plot(oscar_years, oscar_revenue)
plt.title("Oscar Movies Revenue in each Year")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")
plt.show()
```

```python
# Plot multiple
# Non-Oscar Winner Movies
non_oscar_movies = ["Slumdog Millionaire","Avatar","Inception","Hugo","Lincoln"]
non_oscar_years = [2008, 2009, 2010, 2011, 2012]
non_oscar_revenue = [378, 2788, 829, 185, 275]  # in $M

plt.plot(oscar_years, oscar_revenue, ">--r", label="Oscar Winners") #fmt = '[marker][line][color]
plt.plot(oscar_years, non_oscar_revenue, color="blue", linestyle=":",marker="o", linewidth=3, label=" Non Oscar Winners")

plt.title("Oscar winner vs Non Oscar winners")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")
plt.legend()
plt.show()

```

```python
plt.plot(oscar_years, oscar_revenue,  label="Oscar Winners") #fmt = '[marker][line][color]
plt.plot(oscar_years, non_oscar_revenue,  label=" Non Oscar Winners")

plt.title("Oscar winner vs Non Oscar winners")
plt.xlabel("Years")
plt.ylabel("Revenue (in $M)")
plt.legend()
print(plt.style.available) # see available style
plt.style.use("dark_background")

plt.grid()  # show with grid
 
plt.show()
```

```python
## Bar chart

oscar_movies = [
    "The Dark Knight", 
    "The Hurt Locker",  
    "The King's Speech", 
    "The Artist",     
    "Argo"
]
oscar_years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

plt.bar(oscar_years, oscar_revenue, color="green")
plt.title("Revenue in each year for Oscar movies")
plt.xlabel("Year")
plt.ylabel("Revenue (in $M)")

for i in range(len(oscar_years)):
    plt.text(oscar_years[i], oscar_revenue[i]+20, str(oscar_revenue[i]), ha="center")
    # add label in bar x.text(xpos yPos, text, where)

plt.ylim(0, max(oscar_revenue)+100)  # add graph limit where from start and where end
plt.show()
```

```python
# multiple datasets on bar chart
import numpy as np
oscar_movies = [
    "The Dark Knight", 
    "The Hurt Locker",  
    "The King's Speech", 
    "The Artist",     
    "Argo"
]
oscar_revenue = [1005, 170, 427, 133, 232]  # in $M

non_oscar_movies = ["Slumdog Millionaire","Avatar","Inception","Hugo","Lincoln"]
non_oscar_revenue = [378, 2788, 829, 185, 275]  # in $M

years = [2008, 2009, 2010, 2011, 2012]

x = np.arange(len(years)) # array of index length of year
width = 0.4

plt.bar(x-width/2, oscar_revenue, width, label="Oscar Movies") #shifting left
plt.bar(x+width/2, non_oscar_revenue, width, label="Non Oscar Movies") #shifting right

plt.title("Oscar vs Non Oscar Movies Revenue")
plt.xlabel("Year")
plt.ylabel("Revenue ($M)")
plt.legend()

plt.xticks(x, years) #map with np array x with years
```

```python
# Scatter plot
# Scatter Plots
people = ["Person A", "Person B", "Person C", "Person D", "Person E", 
          "Person F", "Person G", "Person H", "Person I", "Person J"]
age = [22, 25, 30, 35, 40, 45, 50, 55, 60, 65]
blood_pressure = [110, 115, 120, 122, 125, 130, 135, 123, 145, 150]
colors =  ["green" if x < 135 else "red" for x in blood_pressure]

plt.scatter(age, blood_pressure, s = blood_pressure, cmap="OrRd", c = blood_pressure) # colors are can be by cmap by value.
                                                                                    #or we can define by color=colors (customize by logic)
plt.title("Age vs BP") 
plt.xlabel("Age")
plt.ylabel("Blood Pressure")

plt.colorbar(label="BP")
plt.grid()

# add Annotation
for i in range(len(people)):
    plt.annotate(people[i], xy=(age[i], blood_pressure[i]), xytext=(age[i]+1, blood_pressure[i]))

plt.xlim(min(age), max(age)+10)
plt.ylim(min(blood_pressure), max(blood_pressure)+5)
plt.show()
```

```python
# Multiple scatter plots
cities = ["City A", "City B", "City C", "City D", "City E"]

# Winter: Temperature (C) vs Humidity (%)
winter_temp = [5, 2, 10, 0, 7]
winter_humidity = [80, 75, 65, 85, 70]

# Summer: Temperature (C) vs Humidity (%)
summer_temp = [25, 30, 28, 35, 27]
summer_humidity = [60, 50, 55, 45, 65]


plt.scatter(winter_temp, winter_humidity, label="winter" )
plt.scatter(summer_temp, summer_humidity, label="Summer")

plt.title("Summer vs Winter data")
plt.xlabel("Temperature (C)")
plt.ylabel("Humidity (%)")
plt.legend()
```

```python
# Pie Charts
expenses = ["Salaries", "Rent", "Marketing", "R&D", "Miscellaneous"]
amounts = [500, 150, 120, 100, 50]

explode = [0,0,0,0.2,0]
plt.pie(amounts, labels=expenses, autopct="%1.1f%%", 
       wedgeprops = {"edgecolor": "black",
                     "linewidth": 1,
                     "linestyle": "--",
                     "alpha" : 1,
                     "fill" : True
                     
                    },

        explode = explode,
        startangle=90,
        shadow=True
       )
plt.show()
```
