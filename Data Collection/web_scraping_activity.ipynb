```python
import requests
from bs4 import BeautifulSoup
```

```python
page_count = 1

while True:
    URL = f"https://quotes.toscrape.com/page/{page_count}/"
    res = requests.get(URL)

    soup = BeautifulSoup(res.text, "lxml")
    quotes = soup.select("div.quote")

    if not quotes:
        print("No valid Pages are anymore...")
        break

    with open(f"scraped_data/quotes{page_count}.html", "w", encoding='utf-8') as f:
        f.write(res.text)
        print(f"Downloaded data from page {page_count}")
    page_count = page_count + 1
```

```python
# extract data 
life_quotes = []

i = 1
while True:
    try:
        with open(f"scraped_data/quotes{i}.html", "r", encoding="utf-8") as f:
            html_content = f.read()
    except FileNotFoundError:
        break  # stop when the file doesn't exist

    soup = BeautifulSoup(html_content, "lxml")
    all_quotes = soup.select("div.quote")

    for q in all_quotes:
        all_tags = [tag.get_text() for tag in q.select(".tags .tag")]

        if "life" in all_tags:
            text = q.select_one("span.text").get_text()
            author = q.select_one("small.author").get_text()
            life_quotes.append([text, author])

    i += 1

print(life_quotes)
```

```python
import pandas as pd
df = pd.DataFrame(life_quotes, columns=["Quotes", "Author"])
df
```
