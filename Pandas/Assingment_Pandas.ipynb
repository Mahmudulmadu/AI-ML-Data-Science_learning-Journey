```python
df.groupby("species").agg(
    max_petal_width = ("petal_width", "max"),
    avg_sepal_length = ("sepal_length" , "mean"),
    std_sepal_length = ("sepal_width" , "std")
)
```

```python
df["petal_ration"] =  df.apply(lambda x : x["petal_length"]/x["petal_width"] , axis=1 )
```

```python
df.groupby("species")["petal_ration"].mean()
```
