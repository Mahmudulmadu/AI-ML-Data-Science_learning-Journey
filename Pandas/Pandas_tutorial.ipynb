```python
# Cleaning Data
df = pd.read_csv("raw_data.csv") 

# Handle Missing values
df.isnull()            # True for null values (NaN)
df.isnull().sum()      # counts missing vals per column

df.dropna()            # drops rows with missing values
df.dropna(axis = 1)    # drops cols with missing values

df.fillna(0)           # fills NaN with value
df["age"] = df["age"].fillna(df["age"].mean())           # fills column with mean

df.ffill()             # forward fill
df.bfill()             # backward fill
```

```python
# Handle Data Types
df = pd.read_csv("raw_data.csv") 

df.dtypes                                     # dtype of each col
df2 = df.fillna(0).copy()
df2["income"] = df2["income"].astype(int)       # change dtype

# String type
df2["gender"].str.lower()                      # lower case
df2["gender"].str.upper()                      # upper case
df2["gender"].str.capitalize()                 # capitalize

df2["gender"].str.capitalize()                 # capitalize
df2["name"].str.split(" ")                     # splits into lists based on separator value 
type(df2["name"].str.split(" ")[0])            

df2["country"].str.contains("US")
df2["country"].str.contains("india", case=False)
```

```python
df = pd.read_csv("raw_data.csv") 

###Feature Engineering

df2 = df.copy()
df2["tax"] = df2["income"].apply(lambda x : "20%" if x >= 60000 else "10%")  ## apply function higher order because another func is a parameter in this func

## Map
gender_mp = {"Male": "M", "Female":"F", "Unknown": "U"}
df2["gender"] = df2["gender"].map(gender_mp)

#Assign func income increase 10%
df2 = df2.assign(new_income = df2["income"] * 1.1)

df2["country"] = df2["country"].replace("USA" , "US")
df2
```

```python
### Others Transformation Methods

df2.columns = ["Id","Name","Age", "Country", "Gender", "Income","Tax", "New_Income"]
df2.rename(columns={"Income":"Salary"})
df2.rename(index={1: "First"})

df2.sort_values("Income")
#f2.sort_values("Income", ascending=False)
sorted_df = df2.sort_values(["Income", "Age"])

sorted_df.sort_index()
sorted_df.reset_index(drop=True)

#Ranking 
sorted_df["Ranking"] = sorted_df["Income"].rank(ascending=False, method="dense")
sorted_df

#Reorder columns
sorted_df[["Id","Name","Age", "Country", "Gender", "Income", "New_Income", "Tax"]]

 
```

```python
### Group by and Aggregation

df.groupby("country")["income"].mean()
df.groupby("country")["income"].max()
df.groupby("country")["income"].min()

df.groupby("country")["income"].agg(["mean","min", "max"])
df.groupby("country")["income"].aggregate(["mean","min", "max"])

df.groupby("country")["income"].agg(avg_salary="mean", min_salary="min", max_salary = "max")

df.groupby("country").agg({
    "income": "mean",
    "age": "mean"
})

df.groupby("country").agg(
   max_salary=( "income", "max"),
    avg_age = ("age", "mean")
)
```

```python
# Melt and Pivot for Reshaping

df = pd.DataFrame({
    "country" : ["USA", "USA", "Bangladesh", "Bangladesh"],
    "year" : [2020, 2021, 2020, 2021],
    "sales" : [100, 120, 90, 110],
    "profit": [20,25,18,22],
})

df

melted_df = df.melt(
    id_vars = ["country", "year"],
    value_vars = ["sales", "profit"],
    var_name = "metrics",
    value_name = "value"
)

melted_df

original = melted_df.pivot(
    index = ["country", "year"],
    columns = "metrics",
    values = "value"
)
original
```

```python
# Merging & Joining

df_customers= pd.DataFrame({ 
    "customer_id": [1,2,3,4],
    "name":["Adam","Bob","Charlie","Dave"]
}) 
df_orders= pd.DataFrame({ 
    "order_id": [101,102,103,104],
    "customer_id":[2,1,4,5],
    "amount":[250,120,300,180]
})

pd.merge(df_customers, df_orders, on ="customer_id") ### Inner join
pd.merge(df_customers, df_orders, on ="customer_id", how = "left") ## Left join
pd.merge(df_customers, df_orders, on ="customer_id", how = "right") ## right join
pd.merge(df_customers, df_orders, on ="customer_id", how = "outer")  ## Outer Join
```

```python
# Data Concatenation

df1= pd.DataFrame({ 
    "id": [1,2,3],"name":["Adam","Bob","Charlie"]
}) 
df2= pd.DataFrame({ 
    "id": [4,5,6],"name":["David","Eva","Frank"]
})

pd.concat([df1, df2], ignore_index=True)

pd.concat([df1, df2], axis=1)
```
