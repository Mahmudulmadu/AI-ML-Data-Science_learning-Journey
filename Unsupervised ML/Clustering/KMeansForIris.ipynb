```python
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
```

```python
iris = load_iris()
```

```python
X= iris.data
y= iris.target
```

```python
X.shape
```

```python
#Visualize
sns.scatterplot(x=X[:, 0], y=X[:, 2])
```

```python
# Scaling data
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(X)
```

```python
# Optional Dimensionality Reduction using PCA
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

pca_data = pca.fit_transform(X_scaled)
```

```python
# Elbow Method

wcss=[]

for k in range(1,11):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit_predict(pca_data)
    wcss.append(kmeans.inertia_)
```

```python
#plot k vs wcss

sns.lineplot(x=range(1,11), y=wcss, marker='o')
```

```python
#Kmeans

kmeans = KMeans(n_clusters=3, random_state=10)
lebels = kmeans.fit_predict(pca_data)
```

```python
sns.scatterplot(x=pca_data[:, 0], y=pca_data[:, 1], c=lebels)
sns.scatterplot(x=kmeans.cluster_centers_[:, 0], y=kmeans.cluster_centers_[:, 1], marker="X", c="red", s=150)
```
