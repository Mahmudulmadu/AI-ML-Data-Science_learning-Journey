```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.set_theme(style="whitegrid")
sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="smoker",
    size="size",
    style="sex"
)
```

```python
sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    kind="line",
   
)
```

```python
sns.scatterplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="smoker",
    size="size",
    style="sex"
)
```

```python
#lineplot

flights = sns.load_dataset("flights")

sns.lineplot(
    data=flights,
    x="year",
    y="passengers",
    
)
```

```python
flights
```

```python
sns.barplot(
    data = tips,
    x="day",
    y="tip",
    hue="sex"
)
```

```python
sns.boxplot(
    data = tips,
    x="day",
    y="tip",
    hue="sex"
)
```

```python
# Distribution --- Histogram
penguins = sns.load_dataset("penguins")

sns.histplot(
    data=penguins,
    x="body_mass_g",
    bins= 20
)
```

```python
# matrix plots --> Heatmaps
flights_pivot = flights.pivot(index="month", columns="year", values="passengers")

sns.heatmap(
    flights_pivot,
    cmap="coolwarm",
    annot=True,
    fmt="d"
)

plt.title("passengers Heatmap")
plt.tight_layout()
```

```python
# best practices in seaborn

fig, ax =  plt.subplots()

sns.lineplot(
    data=tips,
    x="day",
    y="total_bill",
    hue="sex",
    marker="o",
    ax=ax,
    errorbar=None
)

ax.set_title("Avg bill value for each day")
ax.set_xlabel("Days")
ax.set_ylabel("Total Bill Amount")
ax.grid(False)

fig.tight_layout()
```
