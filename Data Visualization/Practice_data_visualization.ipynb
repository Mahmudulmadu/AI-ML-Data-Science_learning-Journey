```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
data = np.random.normal(loc=70, scale=10, size=200)

mean = np.mean(data)
plt.hist(
    data,
    bins=20,
)
plt.axvline(
    x = mean,
    label ="Mean of scores",
    color = "red",
    linewidth = 2,
    linestyle= "--"
    
)
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.title("Histogram of Scores with Mean Line")
plt.tight_layout()
```

```python
fig, axes = plt.subplots()

sns.histplot(
    data,
    bins=20,
    kde=True,       
    ax=axes         
)

axes.axvline(mean)
axes.set_title("Score Distribution")
axes.set_xlabel("Score")
axes.set_ylabel("Frequency")
fig.tight_layout()
plt.show()
```

```python
tips = sns.load_dataset("tips")
sns.boxplot(x="day", y="total_bill", hue="sex", data=tips, palette="Set2")

plt.title("Total Bill by Day and Sex")
plt.xlabel("Day")
plt.ylabel("Total Bill ($)")
plt.show()
```

```python
pivot_table = tips.pivot_table(
    values="tip",
    index="day",
    columns="time",
    aggfunc="mean"
)

pivot_table
```

```python
plt.figure(figsize=(8, 5))

sns.heatmap(
    pivot_table,
    annot=True,          # show values
    fmt=".2f",           # 2 decimal places
    cmap="YlOrRd",       # yellow  red
    linewidths=0.5
)

plt.title("Average Tip by Day and Time")
plt.ylabel("Day")
plt.xlabel("Time")
plt.show()
```
